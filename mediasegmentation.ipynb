```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 1. Generate Synthetic Media Consumer Dataset
np.random.seed(42)
n_users = 1000

data = {
    'user_id': range(1001, 1001 + n_users),
    'recency_days': np.random.choice([np.random.randint(0, 5), np.random.randint(5, 20), np.random.randint(20, 90)], p=[0.5, 0.3, 0.2], size=n_users),
    'monthly_watch_hours': np.random.choice([np.random.normal(40, 8), np.random.normal(12, 3), np.random.normal(2, 1)], p=[0.4, 0.4, 0.2], size=n_users).clip(0.5, 120),
    'completion_rate': np.random.beta(a=5, b=2, size=n_users),
    'binge_index': np.random.poisson(lam=3, size=n_users),
    'ctv_preference_pct': np.random.uniform(0, 1, size=n_users),
    'monetary_value': np.random.choice([14.99, 9.99, 0.00], p=[0.4, 0.5, 0.1], size=n_users)
}
df = pd.DataFrame(data)

# 2. Feature Selection & Preprocessing
features = ['recency_days', 'monthly_watch_hours', 'completion_rate', 'binge_index', 'ctv_preference_pct', 'monetary_value']
X = df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Clustering (K-Means)
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(X_scaled)

# 4. Evaluate and Profile
sil_avg = silhouette_score(X_scaled, df['cluster'])
print(f"Average Silhouette Score: {sil_avg:.3f}")
```
