```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# 1. Mocking interaction data
data = {
    'user_id': ['User_A', 'User_A', 'User_A', 'User_B', 'User_B', 'User_C', 'User_C', 'User_C', 'User_D', 'User_D'],
    'item_id': ['Movie_1', 'Movie_2', 'Movie_3', 'Movie_1', 'Movie_4', 'Movie_2', 'Movie_3', 'Movie_4', 'Movie_1', 'Movie_3'],
    'rating': [5.0, 3.0, 1.0, 4.0, 2.0, 4.0, 5.0, 4.0, 1.0, 5.0]
}
df = pd.DataFrame(data)

# 2. Pivot data into a User-Item Matrix
user_item_matrix = df.pivot(index='user_id', columns='item_id', values='rating').fillna(0)

# 3. Calculate Item-to-Item Cosine Similarity
item_similarity = cosine_similarity(user_item_matrix.T)
item_similarity_df = pd.DataFrame(item_similarity, index=user_item_matrix.columns, columns=user_item_matrix.columns)

# 4. Logic function to generate top recommendations
def get_item_recommendations(user_id, matrix, similarity_df, top_n=2):
    user_ratings = matrix.loc[user_id]
    known_user_likes = user_ratings[user_ratings > 0]
    scores = pd.Series(dtype=float)
    for item, rating in known_user_likes.items():
        weight_similarity = similarity_df[item] * rating
        scores = scores.add(weight_similarity, fill_value=0)
    scores = scores.drop(known_user_likes.index, errors='ignore')
    return scores.sort_values(ascending=False).head(top_n)

# Example output for User_D
recommendations = get_item_recommendations('User_D', user_item_matrix, item_similarity_df, top_n=1)

```
